```python
from dotenv import load_dotenv
import os
import requests
import pandas as pd
```

```python
dotenv_path = os.path.join(os.path.dirname(os.getcwd()), '.env')
load_dotenv(dotenv_path)
token = os.getenv('DATABURSATIL_TOKEN')
```

```python
# Probando funcionamiento API
inicio = '2011-01-10'
final = '2024-10-31'
emisora_serie = 'FUNO11'
url = f"https://api.databursatil.com/v1/historicos?token={token}&inicio={inicio}&final={final}&emisora_serie={emisora_serie}"
```

```python
r = requests.get(url)
```

```python
# Comprobando que la peticin haya sido exitosa
r.json()
```

```python
df_precios = pd.DataFrame.from_dict(r.json(), orient='index').reset_index()
df_precios.columns = ['Fecha', 'Precio', 'Importe']
df_precios
```

```python
data_path = os.path.join(os.path.dirname(os.getcwd()), 'data', 'raw')
```

```python
df_precios.to_csv(os.path.join(data_path, 'FUNO11.csv'), index=False)
```

```python
codigos_FIBRAs = ['FIBRAPL14', 'FIBRAMQ12', 'TERRA13', 'DANHOS13', 'FMTY14', 'FIHO12']
```

```python
# Repetir el proceso para el resto de FIBRAs

for fibra in codigos_FIBRAs:
    url = f"https://api.databursatil.com/v1/historicos?token={token}&inicio={inicio}&final={final}&emisora_serie={fibra}"
    r = requests.get(url)
    df_precios = pd.DataFrame.from_dict(r.json(), orient='index').reset_index()
    df_precios.columns = ['Fecha', 'Precio', 'Importe']
    df_precios.to_csv(os.path.join(data_path, f'{fibra}.csv'), index=False)
```
